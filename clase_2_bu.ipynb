```python
# Ejercicio 35
"""
años perro <= 2 -> 10.5 años humanos
para los años >2 -> 4 años humanos
error si el numero es negativo
"""

a_perro = int(input("Ingrese la edad de su perro en enteros"))

if(a_perro<0):
  print("Error, el perro no puede tener años negativos")
else:
  if(a_perro<=2):
    print(f"En años humanos el perro tiene {a_perro*10.5} años")
  elif(a_perro>2):
    equ_humano = (2*10.5)+(a_perro-2)*4
    print(f"En años humanos el perro tiene {equ_humano} años ")
```

```python
#Ejercio 60

from random import choice

verdes = [0,-1]
rojos = [1,3, 5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36]
negros = [2,4,6,8,10,11,13,15,17,20,22,24,26,28,29,31,33,35]

todos = verdes + rojos + negros

ruleta = choice(todos)

if(ruleta == -1 ):
  print("Paga 00")
elif(ruleta == 0):
  print("Paga 0")
else:
  print(f"Paga {ruleta}")
  if(ruleta in rojos):
    print("Paga rojos")
  elif(ruleta in negros):
    print("Paga negros")
  
  if(ruleta%2==0):
    print("paga par")
  else:
    print("paga inpar")

  if(ruleta <19):
    print("paga entre 1 y 18")
  else:
    print("paga entre 19 y 36")


```
